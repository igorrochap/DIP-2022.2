```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
chips = cv2.imread("chips.png")

plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(chips, cv2.COLOR_BGR2RGB))
plt.title("Input image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
gray = cv2.cvtColor(chips, cv2.COLOR_BGR2GRAY)

gamma = 2.2
gray = np.power(gray/float(np.max(gray)), gamma)
blur = cv2.GaussianBlur(gray, (5,5), 0)

lower_yellow = np.array([29, 80, 50])
upper_yellow = np.array([40, 255, 255])

lower_green = np.array([40, 50, 50])
upper_green = np.array([80, 255, 255])

lower_red = np.array([160, 50, 50])
upper_red = np.array([180, 255, 255])

lower_orange = np.array([5, 150, 80])
upper_orange = np.array([15, 255, 255])

lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

yellow_mask = cv2.inRange(cv2.cvtColor(chips, cv2.COLOR_BGR2HSV), lower_yellow, upper_yellow)
green_mask = cv2.inRange(cv2.cvtColor(chips, cv2.COLOR_BGR2HSV), lower_green, upper_green)
red_mask = cv2.inRange(cv2.cvtColor(chips, cv2.COLOR_BGR2HSV), lower_red, upper_red)
orange_mask = cv2.inRange(cv2.cvtColor(chips, cv2.COLOR_BGR2HSV), lower_orange, upper_orange)
blue_mask = cv2.inRange(cv2.cvtColor(chips, cv2.COLOR_BGR2HSV), lower_blue, upper_blue)

yellow_chips = cv2.bitwise_and(chips, chips, mask=yellow_mask)
green_chips = cv2.bitwise_and(chips, chips, mask=green_mask)
red_chips = cv2.bitwise_and(chips, chips, mask=red_mask)
orange_chips = cv2.bitwise_and(chips, chips, mask=orange_mask)
blue_chips = cv2.bitwise_and(chips, chips, mask=blue_mask)

plt.figure(figsize=(20,10))

plt.subplot(151)
plt.imshow(cv2.cvtColor(yellow_chips, cv2.COLOR_BGR2RGB))
plt.title("Yellow chips")
plt.xticks([]), plt.yticks([])

plt.subplot(152)
plt.imshow(cv2.cvtColor(green_chips, cv2.COLOR_BGR2RGB))
plt.title("Green chips")
plt.xticks([]), plt.yticks([])

plt.subplot(153)
plt.imshow(cv2.cvtColor(red_chips, cv2.COLOR_BGR2RGB))
plt.title("Red chips")
plt.xticks([]), plt.yticks([])

plt.subplot(154)
plt.imshow(cv2.cvtColor(orange_chips, cv2.COLOR_BGR2RGB))
plt.title("Orange chips")
plt.xticks([]), plt.yticks([])

plt.subplot(155)
plt.imshow(cv2.cvtColor(blue_chips, cv2.COLOR_BGR2RGB))
plt.title("Blue chips")
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
from skimage import exposure
import matplotlib.pyplot as plt
import cv2
```

```python
def show_image(image, name):
  plt.figure(figsize=(10, 10))

  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) 
  plt.title(name)
  plt.xticks([]), plt.yticks([])

  plt.show()
```

```python
src = cv2.imread("empire_state_cloudy.jpg")
ref = cv2.imread("empire_state_sunset.jpg")

is_multi = True if src.shape[-1] > 1 else False
matched = exposure.match_histograms(src, ref, multichannel = is_multi)

show_image(src, "Source")
show_image(ref, "Reference")
show_image(matched, "Matched")
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_image(image, normalized):
  plt.figure(figsize=(10,5))
  plt.subplot(121)
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title("Input image")
  plt.xticks([]), plt.yticks([])

  plt.subplot(122)
  plt.imshow(cv2.cvtColor(normalized, cv2.COLOR_BGR2RGB))
  plt.title("Normalized image")
  plt.xticks([]), plt.yticks([])
  plt.show()

def show_channels(r, g, b):
  plt.figure(figsize=(20,10))
  plt.subplot(142)
  plt.imshow(r, cmap='gray')
  plt.title("Red channel")
  plt.xticks([]), plt.yticks([])

  plt.subplot(143)
  plt.imshow(g, cmap='gray')
  plt.title("Green channel")
  plt.xticks([]), plt.yticks([])

  plt.subplot(144)
  plt.imshow(b, cmap='gray')
  plt.title("Blue channel")
  plt.xticks([]), plt.yticks([])

  plt.show()
```

```python
strawberries = cv2.imread("strawberries.tif")
normalized = cv2.normalize(strawberries, None, 0, 1.0, cv2.NORM_MINMAX, dtype=cv2.CV_32F)

show_image(strawberries, normalized)

print()

b, g, r = cv2.split(normalized)

show_channels(r, g, b)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def detect_defects(image):
  bottle_not_exists = bottle_none(image)
  if bottle_not_exists == 1:
    return [0, 0, 0, 0, 0, 0, 0, bottle_not_exists]
  return [
      content_high(image),
      content_low(image),
      cover_none(image),
      bottle_smashed(image),
      label_white(image),
      label_missplaced(image),
      label_none(image),
      bottle_not_exists,
  ]

def content_high(image):
    height, width, _ = image.shape
    middle_bottle_width = (width//3)
    middle_bottle = image[:, middle_bottle_width:middle_bottle_width*2]

    content_start = 70
    content_end = 165
    middle_bottle_content = middle_bottle[content_start:content_end, :]

    hsv = cv2.cvtColor(middle_bottle_content, cv2.COLOR_BGR2HSV)
    lower_black = np.array([0, 0, 0])
    upper_black = np.array([179, 255, 100])
    mask = cv2.inRange(hsv, lower_black, upper_black)

    total_pixels = middle_bottle_content.shape[0] * middle_bottle_content.shape[1]
    white_pixels = cv2.countNonZero(mask)
    white_percentage = (white_pixels / total_pixels) * 100

    if white_percentage > 29.4:
      return 1
    return 0

def content_low(image):
    height, width, _ = image.shape
    middle_bottle_width = (width//3)
    middle_bottle = image[:, middle_bottle_width:middle_bottle_width*2]

    content_start = 70
    content_end = 180
    middle_bottle_content = middle_bottle[content_start:content_end, :]

    hsv = cv2.cvtColor(middle_bottle_content, cv2.COLOR_BGR2HSV)
    lower_black = np.array([0, 0, 0])
    upper_black = np.array([179, 255, 50])
    mask = cv2.inRange(hsv, lower_black, upper_black)

    total_pixels = middle_bottle_content.shape[0] * middle_bottle_content.shape[1]
    white_pixels = cv2.countNonZero(mask)
    white_percentage = (white_pixels / total_pixels) * 100

    if white_percentage < 6.3:
      return 1
    return 0

def cover_none(image):
    height, width, _ = image.shape
    middle_bottle_width = (width//3)
    middle_bottle = image[:, middle_bottle_width:middle_bottle_width*2]

    cover_start = 0
    cover_end = 70
    middle_bottle_cover = middle_bottle[cover_start:cover_end, :]

    hsv = cv2.cvtColor(middle_bottle_cover, cv2.COLOR_BGR2HSV)
    lower_red = np.array([0, 100, 100])
    upper_red = np.array([10, 255, 255])
    mask = cv2.inRange(hsv, lower_red, upper_red)

    total_pixels = middle_bottle_cover.shape[0] * middle_bottle_cover.shape[1]
    white_pixels = cv2.countNonZero(mask)
    white_percentage = (white_pixels / total_pixels) * 100

    if white_percentage == 0:
      return 1
    return 0

def bottle_smashed(image):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  edges = cv2.Canny(gray, 100, 200)

  height, width = img.shape[:2]
  roi_left = edges[:, :width//3]
  roi_middle = edges[:, width//3:2*width//3]
  roi_right = edges[:, 2*width//3:]
  
  roi_middle_array = np.array(roi_middle)

  if (np.average(roi_middle_array) < 20):
      return 1
  return 0

def label_white(image):
    height, width, _ = image.shape
    middle_bottle_width = (width//3)
    middle_bottle = image[:, middle_bottle_width:middle_bottle_width*2]

    label_start = 170
    label_end = 300
    middle_bottle_label = middle_bottle[label_start:label_end, :]

    hsv = cv2.cvtColor(middle_bottle_label, cv2.COLOR_BGR2HSV)
    lower_white = np.array([0, 0, 200])
    upper_white = np.array([150, 255, 255])
    mask = cv2.inRange(hsv, lower_white, upper_white)

    total_pixels = middle_bottle_label.shape[0] * middle_bottle_label.shape[1]
    white_pixels = cv2.countNonZero(mask)
    white_percentage = (white_pixels / total_pixels) * 100

    if white_percentage >= 78.5:
      return 1
    return 0

def label_missplaced(image):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  edges = cv2.Canny(gray, 100, 200)

  height, width = img.shape[:2]
  roi_left = edges[:, :width//3]
  roi_middle = edges[:, width//3:2*width//3]
  roi_right = edges[:, 2*width//3:]
  
  roi_middle_array = np.array(roi_middle)

  if (np.average(roi_middle_array) < 21):
      return 1
  return 0

def label_none(image):
  resized_image = cv2.resize(image, (300, 300))

  height, width = resized_image.shape[:2]
  piece_width = int(width / 3)

  middle_piece = resized_image[100:, piece_width:piece_width*2]

  gray = cv2.cvtColor(middle_piece, cv2.COLOR_BGR2RGB)

  total_pixels = gray.shape[0] * gray.shape[1]
  black_pixels = np.sum(gray == 0)
  black_percentage = black_pixels / total_pixels

  if black_percentage > 0.2:
      return 1
  return 0

def bottle_none(image):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  edges = cv2.Canny(gray, 100, 200)

  height, width = img.shape[:2]
  roi_left = edges[:, :width//3]
  roi_middle = edges[:, width//3:2*width//3]
  roi_right = edges[:, 2*width//3:]
  
  roi_middle_array = np.array(roi_middle)
  
  if (np.average(roi_middle_array) < 5):
      return 1
  return 0
```

```python
for i in range(77):
  img_name = "train_"+str(i + 1)+".jpg"
  img = cv2.imread("train/"+img_name);
  print(img_name)
  print(detect_defects(img))
  print()
```

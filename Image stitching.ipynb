````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
````

# Implemente ```image stitching``` on images ['ic1.png', 'ic2.png', 'ic3.png', 'ic4.png'] to generate one panorama image 'ic.png'.

````python
stitcher = cv2.Stitcher.create()
images = [cv2.imread("ic1.jpg"), cv2.imread("ic2.jpg"), cv2.imread("ic3.jpg"), cv2.imread("ic4.jpg")]
status, dst = stitcher.stitch(images)

plt.figure(figsize=(10, 10))
plt.title("Panorama")
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()

cv2.imwrite("ic.png", dst)
````

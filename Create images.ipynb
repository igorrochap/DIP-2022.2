```python
import matplotlib.pyplot as plt
import numpy as np
```

# (1) Create the image of a paraboloid with one axis scaled (like an oval paraboloid).

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = (x**2) + (y**2)

fig = plt.figure(figsize=(14, 9))
ax = plt.axes(projection ='3d')
ax.plot_surface(x, y, z, cmap="plasma", linewidth=0, antialiased=True)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

# (2) Create the image of a rotated sin using rotation of coordinates.

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

x, y = np.meshgrid(x, y)

z = (np.sin((np.sqrt((x**2) + (y**2))))) + 1 / 2

fig = plt.figure(figsize=(14, 9))
ax = plt.axes(projection ='3d')
ax.plot_surface(x, y, z, cmap="plasma", linewidth=0, antialiased=True)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

# (3) Create the image of a gaussian.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = 1 / (2 * np.pi) * np.exp(-((x**2) / 2 + (y **2) / 2))

fig = plt.figure(figsize=(14, 9))
ax = plt.axes(projection ='3d')
ax.plot_surface(x, y, z, cmap="plasma", linewidth=0, antialiased=True)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

# (4) Create a function that generates the image of a Gaussian optionally rotated by an angle $\theta$ and with mx, my, sx, sy as input arguments.

```python
def gaussian_rotated(shape = (100, 100), mx = 50, my = 50, sx = 10, sy = 10, theta = 0):
  x, y = np.meshgrid(range(shape[1]), range(shape[0]))
  theta = np.deg2rad(theta)
  
  x -= mx
  y -= my
  
  
  x = x * np.cos(theta) - y * np.sin(theta)
  y = x * np.sin(theta) + y * np.cos(theta)
  
  z = np.exp( - ((x**2)/(2*sx**2) + 
                  (y**2)/(2*sy**2)) )

  fig = plt.figure(figsize=(14, 9))
  ax = plt.axes(projection ='3d')
  ax.plot_surface(x, y, z, cmap="plasma", linewidth=0, antialiased=True)
  ax.set_xlabel("X")
  ax.set_ylabel("Y")
  ax.set_zlabel("Z")

  plt.show()

gaussian_rotated()
gaussian_rotated(mx = 20, my = 50, sx = 10, sy = 10, theta = np.pi / 2)
```

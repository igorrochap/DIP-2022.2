```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def show_image(image, name):
  plt.figure(figsize=(10, 10))

  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) 
  plt.title(name)
  plt.xticks([]), plt.yticks([])

  plt.show()
```

# (1) Crop and flip an image using Numpy array indexing.

```python
image = cv2.imread("images/lights.jpg")
show_image(image, "Original image")

flipped = image[::-1, ::1]
show_image(flipped, "Flipped image with Numpy array indexing")

cropped = image[300:2000, 1000:4000]
show_image(cropped, "Cropped image with Numpy array indexing")

cropped_and_flipped = cropped[::-1, ::1]
show_image(cropped_and_flipped, "Cropped and flipped image with Numpy array indexing")
```

# (2) Implement image translation using Numpy and OpenCV.

```python
image = cv2.imread("images/gandalf.jpg")
show_image(image, "Original image")

tx = -300
ty = -120

translation_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=float)

translated = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
show_image(translated, "Translated image")
```

# (3) Implement image rotation using Numpy and OpenCV.

```python
image = cv2.imread("images/arrow.jpg")
show_image(image, "Original image")

height, width = image.shape[:2]
center = (width // 2, height // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, angle=120, scale=1.0)

rotated = cv2.warpAffine(image, rotation_matrix, (width, height))
show_image(rotated, "Rotated image using Numpy and OpenCV")
```

# (4) Implement image resizing using OpenCV.

```python
image = cv2.imread("images/stark.jpg")
show_image(image, "Original image " + str(image.shape))

height, width = image.shape[:2]

new_height = int(height * 0.8)
new_width = int(width * 0.8)
new_dimensions = (new_width, new_height)

resized = cv2.resize(image, new_dimensions, interpolation = cv2.INTER_AREA)
show_image(resized, "Resized image" + str(resized.shape))
```

# (5) Implement bitwise operations: AND, OR, XOR.

```python
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
plt.imshow(square)
plt.show()

circle = np.zeros((300, 300), np.uint8)
cv2.circle(circle, (150, 150), 115, 255, -1)
plt.imshow(circle)
plt.show()
```

### AND

```python
and_operation = cv2.bitwise_and(square, circle)
plt.imshow(and_operation)
plt.show()
```

### OR

```python
or_operation = cv2.bitwise_or(square, circle)
plt.imshow(or_operation)
plt.show()
```

### XOR

```python
xor_operation = cv2.bitwise_xor(square, circle)
plt.imshow(xor_operation)
plt.show()
```

# (6) Implement the "mask" operation, where a third image 'h' contains only a Region of Interest (ROI -- defined by the second image mask 'g') obtained from the input image 'f'. Note that this Region can be of any shape.

```python
image = cv2.imread("images/wally.jpg")
show_image(image, "Original image")

mask = np.zeros(image.shape[:2], dtype = np.uint8)
cv2.circle(mask, (1885, 1210), 100, 255, -1)
masked = cv2.bitwise_and(image, image, mask = mask)
show_image(masked, "Masked image")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
encoded = cv2.imread("bgr_encoded.png")

plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(encoded, cv2.COLOR_BGR2RGB))
plt.title("Input image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def to_binary(number):
    binary = [int(i) for i in list('{0:0b}'.format(number))]
    for j in range(8 - len(binary)):
        binary.insert(0,0)        
    return binary

def to_decimal(binary):
    x = 0
    for i in range(8):
        x = x + int(binary[i]) * (2**(7 - i))
    return x

def discriminate_bit(bit, channel):
    z = np.zeros([225,225])
    for i in range(225):
        for j in range(225):
            binary = to_binary(channel[i][j])
            for k in range(8):
                if k == bit:
                    binary[k] = binary[k]
                else:
                    binary[k] = 0
            decimal = to_decimal(binary)
            z[i][j] = decimal
    return z
```

## On red channel

```python
b, g, r = cv2.split(encoded)

plt.figure(figsize=(20, 10))
for i in range(1,9):
    plt.subplot(4, 2, i)
    plt.imshow(discriminate_bit(i - 1, r), cmap='gray')
    plt.xticks([]), plt.yticks([])
plt.show()
```

## On green channel

```python
plt.figure(figsize=(20, 10))
for i in range(1,9):
    plt.subplot(4, 2, i)
    plt.imshow(discriminate_bit(i - 1, g), cmap='gray')
    plt.xticks([]), plt.yticks([])
plt.show()
```

## On blue channel

```python
plt.figure(figsize=(20, 10))
for i in range(1,9):
    plt.subplot(4, 2, i)
    plt.imshow(discriminate_bit(i - 1, b), cmap='gray')
    plt.xticks([]), plt.yticks([])
plt.show()
```
